```python
cols = ['FILM', 'Fandango_Stars', 'Fandango_Ratingvalue', 'Fandango_votes', 'Fandango_Difference']
fands = fandango[cols]
fands
```

```python
movie.columns
```

```python
col = ['movie', 'year', 'fandango']
movies = movie[col]
movies
```

```python
movies.sample(20, random_state=1)
```

```python
fanrating = fands[fands['Fandango_votes'] > 30]
fanrating
```

```python
fanrating['year'] = fanrating['FILM'].str[-5:-1] 
fanrating['year']
```

```python
fanrating
```

```python
rating2015 = fanrating[fanrating['year'] == '2015']
rating2015['year'].value_counts()
```

```python
movi2016 = movies[movies['year'] == 2016]
movi2016
```

```python
movi2016['year'].value_counts()
```

```python
import matplotlib.pyplot as plt
from numpy import arange
%matplotlib inline
plt.style.use('fivethirtyeight')

rating2015['Fandango_Stars'].plot.kde(label = '2015', legend = True, figsize = (8,5.5))
movi2016['fandango'].plot.kde(label = '2016', legend = True)

plt.title("Comparing distribution shapes for Fandango's ratings\n(2015 vs 2016)",
          y = 1.07) # the `y` parameter pads the title upward
plt.xlabel('Stars')
plt.xlim(0,5) # because ratings start at 0 and end at 5
plt.xticks(arange(0,5.1,.5))
plt.show()
```

```python
print('2015' + '\n' + '-' * 16) # To help us distinguish between the two tables immediately and
                                # avoid silly mistakes as we read to and fro
rating2015['Fandango_Stars'].value_counts(normalize = True).sort_index() * 100
```

```python
print('2016' + '\n' + '-' * 16)
movi2016['fandango'].value_counts(normalize = True).sort_index() * 100
```

```python
mean2015 = rating2015['Fandango_Stars'].mean()
mean2016 = movi2016['fandango'].mean()

median2015 = rating2015['Fandango_Stars'].median()
median2016 = movi2016['fandango'].median()

mode2015 = rating2015['Fandango_Stars'].mode()[0] # the output of Series.mode() is a bit uncommon
mode2016 = movi2016['fandango'].mode()[0]

combined = pd.DataFrame()
combined['2015'] = [mean2015, median2015, mode2015]
combined['2016'] = [mean2016, median2016, mode2016]
combined.index = ['mean', 'median', 'mode']
combined
```

```python
plt.style.use('fivethirtyeight')
combined['2015'].plot.bar(color = '#0066FF', align = 'center', label = '2015', width = .25)
combined['2016'].plot.bar(color = '#CC0000', align = 'edge', label = '2016', width = .25,
                         rot = 0, figsize = (8,5))

plt.title('Comparing summary statistics: 2015 vs 2016', y = 1.07)
plt.ylim(0,5.5)
plt.yticks(arange(0,5.1,.5))
plt.ylabel('Stars')
plt.legend(framealpha = 0, loc = 'upper center')

```
